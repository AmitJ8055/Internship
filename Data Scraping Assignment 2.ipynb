```python
##  Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
##  have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
##  jobs data.
driver=webdriver.Chrome(r'C:\Users\Amit\Data Science Practice\chromedriver.exe')
driver.get("https://www.naukri.com/")
```

```python
designation=driver.find_element(By.CLASS_NAME,'suggestor-input')
designation.send_keys('Data Analyst')

location=driver.find_element(By.XPATH,'/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input')
location.send_keys('Bangalore')

search=driver.find_element(By.CLASS_NAME,'qsbSubmit')
search.click()
```

```python

job_title1=[]
job_location1=[]
company_name1=[]
experience_required1=[]
title_tags=driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]')
for i in title_tags[0:10]:
    title=i.text
    job_title1.append(title)
    
#//span[@class="ellipsis fleft locWdth"]
location_tag=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft locWdth"]')
for i in location_tag[0:10]:
    location=i.text
    job_location1.append(location)
    
company_tags=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in company_tags[0:10]:
    cname=i.text
    company_name1.append(cname)
    
experience_tags=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft expwdth"]')
for i in experience_tags[0:10]:
    experience=i.text
    experience_required1.append(experience)
    
print(len(job_title1),len(job_location1),len(company_name1),len(experience_required1))

df=pd.DataFrame({'title':job_title1,'location':job_location1,'company':company_name1,'experience':experience_required1})
df
```

```python
## Sun Glass
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome(r'C:\Users\Amit\Data Science Practice\chromedriver.exe')
driver.get("https://www.flipkart.com/")
```

```python
Search=driver.find_element(By.CLASS_NAME,'_3704LK')
Search.send_keys('Sun Glass')
Search_Click=driver.find_element(By.CLASS_NAME,'L0Z3Pu')
Search_Click.click()
```

```python
Brand=[]
ProductDiscription=[]
Price=[]
```

```python
start = 0
end=2
for page in range(start,end):
    Brand_tags=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for i in Brand_tags[0:100]:
        BrandName=i.text
        Brand.append(BrandName)

    Price_tags=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    for i in Price_tags[0:100]:
        PriceT=i.text
        Price.append(PriceT)
    
    Discription_tags=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    for i in Discription_tags[0:100]:
        Discription=i.text
        if(Discription==''):
         ProductDiscription.append('null')
        else:
         ProductDiscription.append(Discription)
   
    Next_Click=driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
    Next_Click.click() 
    time.sleep(5)
```

```python
print(len(Brand),len(ProductDiscription),len(Price))
df=pd.DataFrame({'Band':Brand,'Product Discription':ProductDiscription,'Price':Price})
df
```

```python
## Write a python program to scrape data for Top 1000 Quotes of All Time.
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time 
```

```python
driver=webdriver.Chrome(r'C:\Users\Amit\Data Science Practice\chromedriver.exe')
driver.get('https://www.azquotes.com/')
TopQuotes=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
TopQuotes.click()
```

```python
Author=[]
Quotes=[]
QuoteTags=[]

start = 0
end=10
p=1
for page in range(start,end):
    Quotes_tags=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in Quotes_tags[0:1000]:
        Quot=i.text
        Quotes.append(Quot)
        
    Author_tags=driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in Author_tags[0:1000]:
        Aut=i.text
        Author.append(Aut)
        
    QT_tags=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in QT_tags[0:1000]:
        QT=i.text
        QuoteTags.append(QT)
    
    Next_Click=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[4]/li[12]/a')
    Next_Click.click() 
    time.sleep(2)
    #/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[4]/li[12]/a
    #//a[@href="top_quotes.html?p=',p+1,'"]
```

```python
print(len(Author),len(Quotes),len(QuoteTags))

df=pd.DataFrame({'Author':Author,'Quotes':Quotes,'Quote Tags':QuoteTags})
df
```
